```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('flood_risk_ml_dataset.csv')
df.head()
```

```python
# modifiying the risk category using risk score
def categorize_risk(score):
    if score >= 0 and score < 0.5:
        return 'Low Risk'
    elif score >= 0.5 and score < 0.75:
        return 'Moderate Risk'
    elif score >= 0.75 and score < 1.0:
        return 'High Risk'
    else:
        return 'Very High Risk'
    
df['risk_category'] = df['flood_risk_score'].apply(categorize_risk)
df.head()
```

```python
df['risk_category'].value_counts()
```

```python
#saving the modified dataset    
df.to_csv('flood_risk_dataset_modified.csv', index=False)
```
